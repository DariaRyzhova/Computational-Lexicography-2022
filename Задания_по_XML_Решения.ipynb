## Задание 1

Из файла 'rnc_gostin.xml' извлеките все наречия, т.е. все словоформы, у которых в грамматическом разборе присутствует тег 'ADV'.

````python
tree = ET.parse('./sample_data/rnc_gostin.xml')
root = tree.getroot()
print (root.tag)
````

````python
for ana in root.iter('ana'):
    gram = ana.get('gr')
    if gram[:3] == 'ADV':
        print (ana.tail) # Если хотим получить словоформы
        #print (ana.get('lex')) # Если хотим получить леммы
````

## Задание 2

Из файла 'rnc_gostin.xml' извлеките все текстовые данные, т.е. распечатайте сам текст со знаками препинания без каких-либо элементов разметки.

````python
for p in root.findall('p'):
    if p.text:
        print (p.text)
    string = ''
    for se in p.findall('se'):
        if se.text:
            #print (se.text.strip())
            string += se.text.strip()
        for w in se.findall('w'):
            for ana in w.findall('ana'):
                if ana.tail:
                    #print (ana.tail)
                    string += ana.tail.strip() # вытаскиваем слова
            if w.tail:
                #print (w.tail)
                string += w.tail.strip() + ' ' # вытаскиваем знаки препинания
    print (string)
````

То же самое можно сделать намного проще, используя метод itertext(), который возращает все текстовые фрагменты из заданного элемента и из всех элементов, находящихся внутри него.

````python
for p in root.findall('p'):
    paragraph = ('').join(p.itertext())
    print (paragraph)
````

## Задание 3

Возьмите файл 'eng-rus.tei'. Напишите программу, которая на вход получает английское слово, а на выходе дает его переводы на английский язык, разбитые по значениям. Например:

American

1. американский
2. американец


````python
tree = ET.parse('eng-rus.tei')
root = tree.getroot()
print (root.tag, root.attrib, root.text)
````

````python
eng_word = input('Какое слово вы хотите перевести? ')
for entry in root.iter('entry'):
    form = entry.find('form')
    orth = form.find('orth')
    if orth.text == eng_word:
        for sense in entry.iter('sense'):
            if len(sense.attrib) != 0:
                print (sense.get('n'))
            for quote in sense.iter('quote'):
                print (quote.text)
````

## Задание 4

Переведите словарь из файла 'en-ru-test.txt' в XML-формат следующего вида: 

```xml
<body>
  <entry>
    <form lang="English">American</form>
    <sense>американский</sense>
    <sense>американец</sense>
    <sense>американка</sense>
  </entry>
  ...
</body>
```

Запишите результат в файл.

````python
import codecs
````

````python
dict = {}
with codecs.open('en-ru-test.txt', 'r', 'utf-8') as f_dict:
    for line in f_dict:
        line = line.strip().split('\t')
        if line[0] not in dict:
            dict[line[0]] = []
        dict[line[0]].append(line[2])
````

````python
body = ET.Element('body')
for word in dict:
    entry = ET.SubElement(body, 'entry')
    entry.tail = '\n'
    form = ET.SubElement(entry, 'form')
    form.text = word
    form.attrib = {'lang' : 'English'}
    for trans in dict[word]:
        sense = ET.SubElement(entry, 'sense')
        sense.text = trans
````

````python
tree = ET.ElementTree(body)
tree.write('en-ru.xml', encoding = 'utf-8', xml_declaration = 'True')
````
